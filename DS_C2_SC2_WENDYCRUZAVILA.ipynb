```python
## a. Extraer la información del archivo
vacunasCovid= pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/RetoNumpyPandas/country_vaccinations.csv")
vacunasCovid
```

```python
##  b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones 
##puedes realizar con cada una de ellas, asegurandote que las columnas con fechas sean del tipo datetime64
vacunasCovid['date'] = pd.to_datetime(vacunasCovid['date'])
vacunasCovid.info()
```

```python
## c. Determinar la cantidad de vacunas aplicadas de cada compañía (en base a como lo 
## reporta cada país en la columna vaccines)

vacunasCovidJonhson=vacunasCovid[vacunasCovid['vaccines'].str.contains('Johnson&Johnson',case=False)]
print('Johnson&Johnson:', vacunasCovidJonhson['vaccines'].count())
vacunasCovidModerna=vacunasCovid[vacunasCovid['vaccines'].str.contains('Moderna',case=False)]
print('Moderna:', vacunasCovidModerna['vaccines'].count())
vacunasCovidZF2001=vacunasCovid[vacunasCovid['vaccines'].str.contains('ZF2001',case=False)]
print('ZF200:', vacunasCovidZF2001['vaccines'].count())
vacunasCovidOxford=vacunasCovid[vacunasCovid['vaccines'].str.contains('Oxford/AstraZeneca',case=False)]
print('Oxford/AstraZeneca:', vacunasCovidOxford['vaccines'].count())
vacunasCovidPfizer=vacunasCovid[vacunasCovid['vaccines'].str.contains('Pfizer/BioNTech',case=False)]
print('Pfizer/BioNTech:', vacunasCovidPfizer['vaccines'].count())
vacunasCovidSinopharm=vacunasCovid[vacunasCovid['vaccines'].str.contains('Sinopharm/Beijing',case=False)]
print('Sinopharm/Beijing:', vacunasCovidSinopharm['vaccines'].count())
vacunasCovidSinovac=vacunasCovid[vacunasCovid['vaccines'].str.contains('Sinovac',case=False)]
print('Sinovac:', vacunasCovidSinovac['vaccines'].count())
vacunasCovidSputnik=vacunasCovid[vacunasCovid['vaccines'].str.contains('Sputnik',case=False)]
print('Sputnik:', vacunasCovidSputnik['vaccines'].count())
vacunasCovidSinovac=vacunasCovid[vacunasCovid['vaccines'].str.contains('Sinovac',case=False)]
print('Sinovac:', vacunasCovidSinovac['vaccines'].count())
vacunasCovidCansino=vacunasCovid[vacunasCovid['vaccines'].str.contains('Cansino',case=False)]
print('Cansino:', vacunasCovidCansino['vaccines'].count())
vacunasCovidWuhan=vacunasCovid[vacunasCovid['vaccines'].str.contains('Sinopharm/Wuhan',case=False)]
print('Sinopharm/Wuhan:', vacunasCovidWuhan['vaccines'].count())
vacunasCovidNovavax=vacunasCovid[vacunasCovid['vaccines'].str.contains('Novavax',case=False)]
print('Novavax:', vacunasCovidNovavax['vaccines'].count())
vacunasCovidLight=vacunasCovid[vacunasCovid['vaccines'].str.contains('Light',case=False)]
print('Light:', vacunasCovidLight['vaccines'].count())
vacunasCovidCovaxin=vacunasCovid[vacunasCovid['vaccines'].str.contains('Covaxin',case=False)]
print('Covaxin:', vacunasCovidCovaxin['vaccines'].count())
```

```python
## d. Obtener la cantidad de vacunas aplicadas en todo el mundo
vacunasCovid['daily_vaccinations'].sum()
```

```python
## e. Calcular el promedio de vacunas aplicadas por país
vacunasCovid[['iso_code', 'daily_vaccinations']].groupby('iso_code').mean()
```

```python
##  f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo
vacunasCovidEnero=vacunasCovid[vacunasCovid['date'].isin(['29/01/2021'])]
vacunasCovidEnero.count()
```

```python
##  g. Crear un dataframe nuevo denoninado conDiferencias que contenga los datos originales 
## y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw
conDiferencias=vacunasCovid
conDiferencias['difference'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias
```

```python
## h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua
vacunasCovid['date'].max()- vacunasCovid['date'].min()
```

```python
## i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y 
## una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter / 
## NOta:el carcater separador es la coma, por eso asi lo hice
conCantidad=vacunasCovid
conCantidad['canVac']= conCantidad['vaccines'].str.split(', ').str.len()
conCantidad
```

```python
## j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.
antes20=vacunasCovid
antes20=antes20[antes20['date'] < '2021/12/20']
antes20['date'].max()
antes20
```

```python
##  k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer
pfizer=vacunasCovid
pfizer=pfizer[pfizer['vaccines'].str.contains('Pfizer',case=False)]
pfizer
```

```python
##  l. Almacenar los dataframes generados (conDiferencias,conCantidad,antes20 y pfizer) en 
## un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una 
## hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter
with pd.ExcelWriter("/content/gdrive/MyDrive/Colab Notebooks/RetoNumpyPandas/resultadosReto.xlsx") as writer:
    pfizer.to_excel(writer, sheet_name="pfizer")  
    conCantidad.to_excel(writer, sheet_name="conCantidad") 
    antes20.to_excel(writer, sheet_name="antes20") 
    conDiferencias.to_excel(writer, sheet_name="conDiferencias") 
```
